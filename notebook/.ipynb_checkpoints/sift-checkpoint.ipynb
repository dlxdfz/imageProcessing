# sift特征(Scale-invariant feature transform)

[官网sift算法说明](https://docs.opencv.org/3.1.0/da/df5/tutorial_py_sift_intro.html)  
[博客](http://blog.csdn.net/abcjennifer/article/details/7639681)

## 主要步骤说明

1. 构建尺度空间：二维高斯卷积核，对原图像做平滑（δ=1.6，k=sqrt(2)，k=2^(1/s)，octave=4，每一层平滑图像个数=2+3=5）  
  - 不同尺度（怎么保证旋转不变性）
- DOG尺度空间：Dog角点检测（是LOG角点的近似）(octave=4, 每层DOG差值图像个数=5-1=4)
- DOG尺度空间极值检测：Dog特征空间极值提取(当前层像素和相邻两层像素做比较，一个像素和26个像素做比较，octave=4，每层极值尺度个数=4-2=2)

- 去除不好的特征点
  - 第三步找到极值点，近似** *Harris角点检测？* **
  - D(x)，找到极点后，评价极点是否稳定。（详见博客，将D(x)泰勒二阶展开，通过拟合二次曲线，去除小于阈值的点）
  - 边缘响应（**什么叫边缘响应**），主曲率（**什么叫主曲率，和hessian矩阵有关**）大于阈值的点删除。
- 特征点描述
  - 128维特征(为每个特征点计算一个方向，依照这个方向做进一步的计算， 利用关键点邻域像素的梯度方向分布特性为每个关键点指定方向参数，使算子具备旋转不变性。)
  


```python
import cv2
import numpy as np
```

```python
img = cv2.imread('../data/lena.jpg')
# 构建sift检测子
sift = cv2.xfeatures2d.SIFT_create()
```

## plt.imshow() && cv2.imread()

opencv默认的imread是以BGR的方式进行存储的,而matplotlib的imshow默认则是以RGB格式展示,所以须对图片的通道进行转换

```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
def cv2ImagePltShow(cv2Img):
    img = cv2.cvtColor(cv2Img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
img = cv2.imread('../data/lena.jpg')
cv2ImagePltShow(img)
```
